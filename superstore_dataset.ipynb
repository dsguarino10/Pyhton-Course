```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configuración general
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Cargar el dataset
df = pd.read_csv("superstore_dataset2012.csv", encoding='latin1', parse_dates=['Order Date', 'Ship Date'])

# Exploración inicial
print("Información general del dataset:")
print(df.info())
print("\nResumen estadístico:")
print(df.describe())
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
# Gráfico univariante con Matplotlib - Histograma de Ventas
plt.figure()
plt.hist(df['Sales'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribución de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.savefig("histograma_ventas.png")
plt.show()
```

```python
# Gráfico univariante con Seaborn - Boxplot de Beneficio por Categoría
plt.figure()
sns.boxplot(x='Category', y='Profit', data=df, palette='Set2')
plt.title("Distribución de Beneficio por Categoría")
plt.xlabel("Categoría")
plt.ylabel("Beneficio")
plt.show()
```

```python
# Gráfico bivariante con Matplotlib - Dispersión Ventas vs Beneficio
plt.figure()
plt.scatter(df['Sales'], df['Profit'], alpha=0.5, color='teal')
plt.title("Relación entre Ventas y Beneficio")
plt.xlabel("Ventas")
plt.ylabel("Beneficio")
plt.grid(True)
plt.show()
```

```python
# Gráfico bivariante con Seaborn - Regplot
plt.figure()
sns.regplot(x='Sales', y='Profit', data=df, scatter_kws={'alpha':0.4}, line_kws={'color':'red'})
plt.title("Regresión entre Ventas y Beneficio")
plt.xlabel("Ventas")
plt.ylabel("Beneficio")
plt.show()
```

```python
# Visualización multivariante con Seaborn - Heatmap de correlación
plt.figure()
corr = df[['Sales', 'Profit', 'Quantity', 'Discount']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Mapa de Calor de Correlaciones")
plt.show()
```

```python
# Subplots con múltiples visualizaciones
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Histograma
axs[0, 0].hist(df['Sales'], bins=30, color='skyblue', edgecolor='black')
axs[0, 0].set_title("Histograma de Ventas")
axs[0, 0].set_xlabel("Ventas")
axs[0, 0].set_ylabel("Frecuencia")

# Boxplot
sns.boxplot(x='Category', y='Profit', data=df, palette='Set2', ax=axs[0, 1])
axs[0, 1].set_title("Boxplot de Beneficio por Categoría")

# Dispersión
axs[1, 0].scatter(df['Sales'], df['Profit'], alpha=0.5, color='teal')
axs[1, 0].set_title("Dispersión Ventas vs Beneficio")
axs[1, 0].set_xlabel("Ventas")
axs[1, 0].set_ylabel("Beneficio")

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", ax=axs[1, 1])
axs[1, 1].set_title("Correlaciones")

plt.suptitle("Visualizaciones del Superstore Dataset 2012", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig("superstore_visualizaciones.png")
plt.show()
```
